````python
#5c

def logic_circuit(A, B, C):
    """
    Implements the logic circuit:
    Y = (A AND C) OR (B AND NOT C)
    
    A, B, and C should be Boolean values (True or False).
    """
    if not all(isinstance(x, bool) for x in [A, B, C]):
        raise ValueError("Inputs must be boolean (True/False).")

    return (A and C) or (B and not C)

# example results with different values of C
print(logic_circuit(True, True, False)) 
print(logic_circuit(True, True, True)) 

# used gpt to assist with creating the logic function
````
